```python
import spacy

nlp = spacy.load("en_core_web_sm")

def extract_names_and_titles(data):
    # Extract names from the 'guest' field
    guest_names = []
    for episode in data:
        if 'guest' in episode:
            guest_names.append(episode['guest'])

    
    # Extract names from the 'description' field using NER
    names_from_description = []
    for episode in data:
        if 'description' in episode:
            doc = nlp(episode['description'])
            for entity in doc.ents:
                if entity.label_ == "PERSON":
                    names_from_description.append(entity.text)
    
    # Extract titles and descriptions
    titles = []
    descriptions = []
    for episode in data:
        titles.append(episode['title'])
        descriptions.append(episode['description'])
    
    return guest_names, names_from_description, titles, descriptions
```

```python
# Call the function to generate the lists
guest_names, names_from_description, titles, descriptions = extract_names_and_titles(data)
```

**Guest Co-Appearance Network:**

* Nodes: Guests.
* Edges: Connect two guests if they have appeared together in an episode.
* Use: To see which guests frequently appear together.

```python
import networkx as nx
import itertools

def create_coappearance_network(guest_names):
    G = nx.Graph()
    
    for names in guest_names:
        # Assuming guests are separated by commas
        guests = names.split(',')
        
        # Remove any extra whitespace
        guests = [guest.strip() for guest in guests if len(guest.strip()) > 1]
        
        # Add edges for every pair of guests
        for guest1, guest2 in itertools.combinations(guests, 2):
            if G.has_edge(guest1, guest2):
                # Increment weight by 1 if edge already exists
                G[guest1][guest2]['weight'] += 1
            else:
                G.add_edge(guest1, guest2, weight=1)
                    
    return G

# Create the network using your guest_names
G = create_coappearance_network(guest_names)

# For visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12))
pos = nx.spring_layout(G, seed=42)  # Seed for reproducible layout
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.show()
```

```python
import networkx as nx

G = nx.Graph()

# Iterate over each episode's guest list
for names in guest_names:
    # Split by comma first
    guests = [name.strip() for name in names.split(',')]
    
    # Check if we need further splitting by "&"
    all_guests = []
    for guest in guests:
        if '&' in guest:
            all_guests.extend([name.strip() for name in guest.split('&')])
        else:
            all_guests.append(guest)
    
    # Create edges for every pair of guests in the episode
    for i in range(len(all_guests)):
        for j in range(i+1, len(all_guests)):
            # If edge already exists, increase the weight by 1
            if G.has_edge(all_guests[i], all_guests[j]):
                G[all_guests[i]][all_guests[j]]['weight'] += 1
            else:
                G.add_edge(all_guests[i], all_guests[j], weight=1)

# Visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, width=[G[u][v]['weight'] for u,v in G.edges()])
plt.title("Guest Co-Appearance Network")
plt.show()
```

**Guest-Topic Network:**

* Nodes: Guests and extracted topics or keywords from descriptions.
* Edges: Connect a guest to a topic if that topic is mentioned in an episode where the guest appeared.
* Use: To see the topics associated with each guest.

```python
import networkx as nx

G = nx.Graph()

# Extract topics from descriptions
def extract_topics(description):
    doc = nlp(description)
    # We'll consider entities of type 'ORG', 'NORP', 'FAC', 'GPE', 'LOC', 'PRODUCT', 'EVENT', 'WORK_OF_ART', 'LAW', 'LANGUAGE' as potential topics
    return [entity.text for entity in doc.ents if entity.label_ in ['ORG', 'NORP', 'FAC', 'GPE', 'LOC', 'PRODUCT', 'EVENT', 'WORK_OF_ART', 'LAW', 'LANGUAGE']]

# Create nodes and edges
for episode in data:
    guests = [name.strip() for name in episode['guest'].split(',')]
    topics = extract_topics(episode['description'])
    
    # For bipartite networks, you often need to specify the node type
    for guest in guests:
        G.add_node(guest, node_type='guest')
        
    for topic in topics:
        G.add_node(topic, node_type='topic')
        
    for guest in guests:
        for topic in topics:
            G.add_edge(guest, topic)

# Visualization
plt.figure(figsize=(12,12))

# Separate the guest and topic nodes for visualization
guest_nodes = [n for n, d in G.nodes(data=True) if d['node_type']=='guest']
topic_nodes = [n for n, d in G.nodes(data=True) if d['node_type']=='topic']

# Use a shell layout with guests in the middle and topics surrounding them
pos = nx.shell_layout(G, [topic_nodes, guest_nodes])

nx.draw_networkx_nodes(G, pos, nodelist=guest_nodes, node_color="skyblue", node_size=500)
nx.draw_networkx_nodes(G, pos, nodelist=topic_nodes, node_color="lightgreen", node_size=500)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, font_size=10)

plt.title("Guest-Topic Network")
plt.show()
```

**Guest Popularity Network:**

* Nodes: Guests.
* Edges: Connect a guest to another based on a metric like the number of views or likes (if you have that data).
* Use: To understand which guests are the most popular or influential.

If two guests appear together multiple times, their edge weight gets updated.

The edge widths in the visualization are scaled for better visuals.

```python
# Check if 'views' field exists in the first episode of the data
if 'views' in data[0]:
    print("The 'views' field exists in the data.")
else:
    print("The 'views' field does not exist in the data.")
```

```python
import networkx as nx

# Create an empty weighted graph
G = nx.Graph()

# Iterate over the data to add nodes for each guest and weighted edges based on views
for episode in data:
    guests = [name.strip() for name in episode['guest'].split(',')]
    
    # Consider every pair of guests in the episode
    for i in range(len(guests)):
        for j in range(i+1, len(guests)):
            guest1, guest2 = guests[i], guests[j]
            
            # Convert views to integer
            views = int(episode['views'].replace(',', ''))

            
            # If both guests are already connected, update the edge weight
            if G.has_edge(guest1, guest2):
                G[guest1][guest2]['weight'] += views
            else:
                G.add_edge(guest1, guest2, weight=views)



# Visualization
plt.figure(figsize=(12,12))
edge_widths = [d['weight'] / 1000000 for _, _, d in G.edges(data=True)] # This scales the view count for better visualization
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color="skyblue", node_size=1000)
nx.draw_networkx_edges(G, pos, width=edge_widths)
nx.draw_networkx_labels(G, pos, font_size=10)

plt.title("Guest Popularity Network")
plt.show()
```

**Now lets do some clustering ideas.**

Community Detection: Identify clusters or communities within the network. 
Use Louvain algorythm to group nodes that are more connected together than with the rest of the network.

```python
# !pip install python-louvain
```

```python
import community as community_louvain

# Use the Louvain algorithm to detect communities
partition = community_louvain.best_partition(G)

# Visualize the communities
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 10))
plt.axis('off')
nx.draw_networkx_nodes(G, pos, node_size=600, cmap=plt.cm.RdYlBu, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.3)
plt.show()

# Print the communities
for community_num, nodes in enumerate(partition):
    print(f"Community {community_num}: {nodes}")
```

Lets use an other algorythm: Girvan-Newman

```python
from networkx.algorithms.community import girvan_newman

# Girvan-Newman algorithm to detect communities
communities = girvan_newman(G)

# Get the first partitioning (you can also explore deeper levels)
first_communities = next(communities)

# Print the communities
for community_num, community in enumerate(first_communities):
    print(f"Community {community_num}: {community}")
```
