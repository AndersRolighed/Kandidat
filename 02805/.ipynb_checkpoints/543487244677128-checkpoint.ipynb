### Part 0: Building the network 

The get request from the wikipedia API with base url"https://en.wikipedia.org/w/api.php?", with some cleaning and processing, out puts a string containing all the text on a given wikipedia page. A wikipedia page has a lot of hyper links that direct a user to a different site if the link it clicked. These hyperlinks are presented in the text out put as some text between two sets of square brackets. So, by finding all instances of text enclosed by two pairs of square brakcets, we can get a list of links on a given page. Comparing this list to known rappers, we can trivially see if a rapper is mentioned as a hyper link on another rappers page.

The regular expression used to captures these hyperlinks takes the form `\[\[(.+?)\]\]` where `(.+?)` finds and captures everything enclosed by `\[\[  \]\]`


First, we scrape wikipedia for lists of west coast and east coast rappers

```python
# west coast
# define wiki page

site = pywikibot.Site("en", "wikipedia")
category = pywikibot.Category(site, 'West_Coast_hip_hop_musicians')

articles = category.articles()

# convert to list of West Coast artists
westcoast = []

# loop over all articles and ajdust format
for article in list(articles):
    
    link = str(article)
    page = str(re.findall(r'\[\[(.+?)\]\]',link))
    artist = str(page)[5:-2]
    westcoast.append(artist)

print(westcoast)
# east coast

site = pywikibot.Site("en", "wikipedia")
category = pywikibot.Category(site, 'East_Coast_hip_hop_musicians')

articles = category.articles()

# convert to list of West Coast artists
eastcoast = []

# loop over all articles and ajdust format
for article in list(articles):
    
    link = str(article)
    page = str(re.findall(r'\[\[(.+?)\]\]',link))
    artist = str(page)[5:-2]
    eastcoast.append(artist)

print(eastcoast)
```

For each rapper, we scrape the corresponding wikipage and collect all hyper links. We compare the list of hyper links to the list of rappers and create a link between the rappers if a rapper is mentioned in a rappers page. Specifically, for the directed case, we define the direction from the rappers page, to mentioned rappers.

```python
# initialise directional graph
G = nx.DiGraph()

# add all west coast nodes
G.add_nodes_from(westcoast, type='west')
G.add_nodes_from(eastcoast, type='east')

# define all rappers
rappers = eastcoast + westcoast

# initialise scraping
baseurl = "https://en.wikipedia.org/w/api.php?"

# loop over all rappers
for rapper in rappers:

    # define params
    params = {
    'action': 'query',
    'prop': 'revisions',
    'rvprop': 'content',
    'format': 'json',
    'titles': rapper
    }
    # scrape
    wikiraw = requests.get(baseurl, params=params)
    wikijson = wikiraw.json()
    pagenumber = list(wikijson['query']['pages'].keys())[0]
    wikitext = wikijson['query']['pages'][pagenumber]['revisions'][0]['*']
    
    # compute number of words in the wiki page
    numberOfWords = len(re.findall(r'\w+(-\w+)*',wikitext))
    G.nodes[rapper]['length'] = numberOfWords

    # find all pages that are linked
    pages = re.findall(r'\[\[(.+?)\]\]',wikitext)

    # find references to other rappers and add an edge in the network
    for page in pages:
        if page in rappers:
            
            # find direction of edge
            if rapper in eastcoast:
                raptype = 'east'
            else:
                raptype = 'west'
            if page in eastcoast:
                pagetype = 'east'
            else:
                pagetype = 'west'

            label = raptype + ' to ' + pagetype 

            # add directed edge with direction as attribute
            G.add_edge(rapper,page, direction=label)

# remove isolated nodes
G.remove_nodes_from(list(nx.isolates(G)))

# extract the largest component
Gcc = sorted(nx.weakly_connected_components(G),key=len,reverse=True)
G0 = G.subgraph(Gcc[0])
G = G0
#  convert to undirected graph
N = G0.to_undirected()
```

Evidently, both in and out degree distributions seem to approximate a power law distribution (evident by the approximate linear log log plot). This means that there are exponentially more nodes with low degrees that higher degrees.

The in and out distributions are significantly different from eachother. This is due to the fact that just because rapper A is mentioned in rapper B's page, does not necessarily mean that rapper B is mentioned in rapper A's page. This tendency is only increased if there is a large difference in popularity and fame, between the rappers in question. If a smaller rapper had a connection to a more famous rapper, it is for sure going to be mentioned in the wikipedia page, but it is not likely that the famous rapper will have mentioned the lesser rapper as there are likely more important and impactful things to write about on their page. 

We can exponent that best approximates the above degree distribution power laws

```python
indegrees =  sorted((d for n, d in G.in_degree()), reverse=True)
outdegrees =  sorted((d for n, d in G.out_degree()), reverse=True)

print(f'the in degree exponent : {powerlaw.Fit(indegrees,suppress_output=True).alpha:.2f}')
```

```python
print(f'the in degree exponent : {powerlaw.Fit(outdegrees,suppress_output=True).alpha :.2f}')
```

In the regime of $ \gamma > 3 $ we note that the mean node to node distance scales by $ \ln N $, although hubs exist, they are not suffeciently large for their to be a significant impact to the mean distance.

The smaller the explonent, the more the hubs are impacting the mean distance. This is also what we see in the degree distribution histograms shown below. The out degrees behave more like random networks, when compared to the in degree distribution.

We generate random ER network with similar probabilty of connection

```python
n = len(G.nodes())
k = np.mean(sorted((d for n, d in N.degree()), reverse=True))
p = k/(n-1)

# generate ER network
ER = nx.fast_gnp_random_graph(n=n,p=p)
```

Now we plot the degree distribution of the ER network

```python
degrees = sorted((d for n, d in ER.degree()), reverse=True)

# bin the degrees
minsample = min(degrees)
maxsample = max(degrees)

# create binning vector
v = np.arange(minsample,maxsample + 1)

# bin using the bins described
counts, bins = np.histogram(degrees,bins=v)

# generate plot
fig, axs = fig, axs = plt.subplots(2, 1, tight_layout=True)
axs[0].plot(bins[:-1],counts, color='olivedrab',marker='x',linestyle = 'None')
axs[0].set_ylabel('count')
axs[0].title.set_text('ER degrees')
axs[1].loglog(counts, color='olivedrab',marker='x',linestyle = 'None')
axs[1].set_ylabel('count')
axs[1].set_xlabel('degree')

plt.show()
```

With the random ER network we see that the frequency peaks at around the mean degrees where it seems that it is equally likely to have degree less as more than the mean, unlike the in and out distributions of the rapper network. This also means that the ER network does not follow the power law that the rapper network approximates, see very un linear log log plot of the same distribution. This means that in the ER network, the mean distance is seemingly not dependent on a few highly connected nodes that function as hubs.

We read west and east coast .txt files and tokenize them. We remove punctuation and set everything to lower case. We remove stop words and lemmatize them using built in NLTK functions.

```python
eastcoastfile = open('data/eastcoast.txt', encoding='utf-8')
eastcoast = eastcoastfile.read()

westcoastfile = open('data/westcoast.txt', encoding='utf-8')
westcoast = westcoastfile.read()

# tokenize
eastcoast = nltk.word_tokenize(eastcoast)
westcoast = nltk.word_tokenize(westcoast)

# remove punctuation and set lower case
eastcoast = [word.lower() for word in eastcoast if word.isalpha()]
westcoast = [word.lower() for word in westcoast if word.isalpha()]

# remove stop words
stop_words = set(stopwords.words('english'))

eastcoast = [word for word in eastcoast if not word in stop_words]
westcoast = [word for word in westcoast if not word in stop_words]

# lemmatize
wnl = WordNetLemmatizer()

eastcoast = [wnl.lemmatize(word) for word in eastcoast]
westcoast = [wnl.lemmatize(word) for word in westcoast]
```

We generate TF's

```python
# generate TF's
eastTF = nltk.FreqDist(eastcoast)
westTF = nltk.FreqDist(westcoast)
```

We use the TF-TR method described in the week 7 exercises. The below is an implementation of that.

```python
# TF-TR generation
# initialise
eastweights = {}
westweights = {}

eastTFIF = {}
westTFIF = {}

# define all distinct words
allwords = list(set(westcoast+eastcoast))
# define c
c = 5

for word in allwords:
    try:
        TFe = eastTF[word]
    except:
        TFe = 0
    try:
        TFw = westTF[word]
    except:
        TFw = 0

    ww = TFw / (TFe + c)
    ew = TFe / (TFw + c)

    eastweights[word] = ew
    westweights[word] = ww


for word in list(set(eastcoast)):
    eastTFIF[word] = eastTF[word]*eastweights[word]

for word in list(set(westcoast)):
    westTFIF[word] = westTF[word]*westweights[word]


eastTFIF =  dict(sorted(eastTFIF.items(), key=lambda item: item[1],reverse=True))
westTFIF =  dict(sorted(westTFIF.items(), key=lambda item: item[1],reverse=True))
```

We prepare data for the word cloud. This means that we make strings that include a word an integer rounded TF-IF weight times.

```python
# prepare data for word cloud
easttext = ''
westtext = ''

for word in list(set(eastcoast)):
    count = int(eastTFIF[word])

    for i in range(count):
        easttext = easttext + ' ' + word

print('done with east coast')

for word in list(set(westcoast)):
    count = int(westTFIF[word])

    for i in range(count):
        westtext = westtext + ' ' + word

print('done with west coast')
```

### Part 3: Communities


Network modularity is a measure for how modular a network is. This means that if a network has groupings of nodes that are connected to eachother, and not so connected to other groupings, then the network is very modular.

```python
# undirected network is referred to as N
N
```

We use the louvain algorithm to find partitions. The algorithm works by computing the modularity of the network assuming that all nodes belong their own partition. The modularity is then computed for cases when nodes belong to each others partition, and keep then change if the total modularity of the network is increased until a maximum is reached.

```python
partition = community.community_louvain.best_partition(N)
print(f'the algorithm found {max(partition.values())+1} distinct partitions' ) 
```

```python
comms = []

for i in range(max(partition.values())+1):
    d = dict((k, v) for k, v in partition.items() if v==i)
    comms.append(d.keys())
```

```python
print(f'the modularity of this given partition is {nx.community.modularity(N, comms):.2f}')
```

```python
sizes = []
for i in comms:
    sizes.append(len(i))

plt.bar(range(max(partition.values())+1),sizes,color='olivedrab')
plt.xlabel('Partition')
plt.ylabel('Size')
plt.title('Size of each partitioned group of ndoes')
plt.show()
```

We see that not all partitions are of equal size, which could mean that their are some very few rappers that are connected to each other but not very connected with the rest of the rapper network. This could represent some small niche sub genres or communities.

### Part 4: Sentiment of communities

We define a function that can compute a measure of sentiment given 

```python
# read world list
wordlist = pd.read_csv('data/wordlist.txt', sep='\t')
words = list(wordlist['word'])
scores = list(wordlist['happiness_average'])

# convert to dict
global scoresdict
scoresdict = {}

for i, word in enumerate(words):
    scoresdict[word] = scores[i]

# function for computing the sentiment of a list of tokens
def sentiment(tokens):

    # only iterested in tokens that are in the defined world list
    tokensWithScore = [x for x in tokens if x in list(scoresdict.keys())]

    # compute freq dist
    FD = nltk.FreqDist(tokensWithScore)
    # total sum
    total = sum(FD.values())

    # initialise
    result = 0

    # compute weighted mean
    for token in list(set(tokensWithScore)):
        result += (FD[token]*scoresdict[token])/total

    return result
```

```python
# compute sentiment for each rapper
stop_words = set(stopwords.words('english'))
wnl = WordNetLemmatizer()

sentiments = {}

# loop over all rappers in scope
for rapper in rappers:

    # open .txt file
    textfile = open(f'data/{rapper}.txt', encoding='utf-8')
    text = textfile.read()
    
    # tokenize
    text = nltk.word_tokenize(text)

    # remove punctuation and set to lower case
    text = [word.lower() for word in text if word.isalpha()]

    # remove stop words
    text = [word for word in text if not word in stop_words]

    # lemmatizr
    text = [wnl.lemmatize(word) for word in text]

    # save sentiment
    sentiments[rapper] = sentiment(text)
```
