# Part 0: Answer
Once we had downloaded each page we applied the function below to the text. This function finds all the links in the text by using a regex-pattern that matches the structure of a link in a Wikipedia page. The structure of a link is either **\[\[link\]\]** or **\[\[link|text\]\]** where **link** is the part of interest since it is in the same format as the **WikipediaPageName** feature of the dataset of all the rappers. The **link** part is extracted by splitting the regex-match by **|** and then taking the first element.
Once the links are found we check whether the **link** is in the list of rappers, if they are we save the link such that it can be used to build the edges of the network.

Now we will walk through the regex pattern we used to extract the links. Our pattern looks like this **``r"\[\[([^\]]+)\]\]"``**, a breakdown of the different components is as follows,
* The pattern is contained in double-quotes with an `r` preceding it, this makes it a raw string which allows us to use `\` without directly without having to type it twice
* The `\` is used to escape certain characters in the pattern in our case this concerns `[` and `]`, these are normally used to indicate a set of characters in regex but in our case we want to indicate that the link must start and end with two of the `[` and `]` respectively, in order to do so we add escape the square brackets
* After the initial square brackets we see a `(` which encloses the inner contents of the pattern. This syntax is used to define a group which is useful when using the results from the `.findall` operation. By forming a group each match is returned without the surrounding square brackets which makes extracting the link easier.
* Finally, we define the characters that lie between the square brackets, here we want to allow an arbitrary length of characters this is done by using `+` which means that one or more characters these characters are defined in the set of characters preceding the `+` which is set to be all characters but `]`, which is not as parts of the link since it must be part of the closing sequence, the `^`-sign means not which in this case excludes `]`.

```python
import re

import pandas


def find_links(text):
    matches = re.findall(r"\[\[([^\]]+)\]\]", text)
    return list(map(lambda x : x.split("|")[0], matches))
```

First we load in all the text files from the dataset supplied in the course. The texts are loaded into separate variables one for each coast

```python
import os

import nltk
from nltk.tokenize import RegexpTokenizer


separate_east_coast_texts = []
for file in os.listdir("eastcoasttexts/eastcoasttexts"):
    with open("eastcoasttexts/eastcoasttexts/"+file, "r", encoding="utf-8") as f:
        text = f.read()
        separate_east_coast_texts.append((file.split(".")[0], text))

separate_west_coast_texts = []
for file in os.listdir("westcoasttexts/westcoasttexts"):
    with open("westcoasttexts/westcoasttexts/"+file, "r", encoding="utf-8") as f:
        text = f.read()
        separate_west_coast_texts.append((file.split(".")[0], text))
```

Then we tokenize our text. The regex-patter specified in the first line matches all character sequences from these sequences we make sure that they are all in lowercase letters such that we have a tokenized dataset to work on.

```python
non_punctuation_tokenizer = RegexpTokenizer(r"\w+")
east_rapper_tokens = [(x[0],  [w.lower() for w in non_punctuation_tokenizer.tokenize(x[1])]) for x in separate_east_coast_texts]
west_rapper_tokens = [(x[0],  [w.lower() for w in non_punctuation_tokenizer.tokenize(x[1])]) for x in separate_west_coast_texts]
```

```python
east_tokens = [token for rapper in east_rapper_tokens for token in rapper[1]]
west_tokens = [token for rapper in west_rapper_tokens for token in rapper[1]]
```

Now we create histograms for each word in the respective texts

```python
east_frequencies = nltk.FreqDist(east_tokens)
west_frequencies = nltk.FreqDist(west_tokens)
```

This is the function for calculating the weight of a word. Note that $ c=1 $ is used as a default higher values of $ c $ makes the formula more sensitive to the home count since the denominator is larger by default

```python

def word_weight(home_count, foreign_count, c = 1):
    return home_count/(foreign_count+c)
```

Here the TF-TR values are calculated for each coast by using the former function and multiplying the value with the word frequency

```python
west_TF_TR = {}
for word in set(west_tokens):
    west_TF_TR[word] = word_weight(west_frequencies[word], east_frequencies[word])*west_frequencies[word]
```

```python
east_TF_TR = {}
for word in set(east_tokens):
    east_TF_TR[word] = word_weight(east_frequencies[word], west_frequencies[word])*east_frequencies[word]
```

Now we can create the wordclouds, we do this by taking each token and creating a pseudo-text where each word occurs TF-TR-times such that the word is weighted appropriately. We omit collocations since the pseudo-text contains every occurrence of each token in sequence which would create some strange results if we did not exclude collocations.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
def draw_word_cloud(tokens, TF_TR):
    text = ""
    for word in set(tokens):
        text += (word + " ")*int(TF_TR[word])
    word_cloud = WordCloud(collocations=False).generate(text)
    plt.imshow(word_cloud)
    plt.axis("off")
    plt.show()
```

The size of the words on this image is determined by their frequency in the pseudo-text, and hence their TF-TR-values. We observe that the name of some rappers seem to be very large in the image which makes sense since these rappers are linked to one coast which makes their name way more frequent in texts of that coast which boost their TF-TR value. On the last image we see words as record and song appearing. This is probably not because these words appear more frequently on one coast than the other but rather because they are common words when writing about rappers. We notice that common words like; the, I and it, does not apper in the word-clouds it appears as if they have been removed from the clouds even though they have high TF-TR values as we will see in the next question.

##### Question: In your own words, explain what the measure “modularity” is, and the intuition behind the formula you use to compute it.
###### Answer: Modularity is a way of describing how well we have partitioned our graph into communities. The modularity-term is best understood by considering the formula which calculates the modularity
$$ M = \sum_{c=1}^{n_c} {[\frac{L_c}{L}- (\frac{k_c}{2L})^2]} $$

Here $n_c$ is the number of partitions we have split our graph into, $L_c$ and $k_c$ are the number of links and the sum of degrees in the $c$'th community respectively.

In a well partitioned graph the communities have many connections to the fellow nodes. An example is a school where each class might constitute a good community since each student knows all the fellow students quite well compared to students of other classes.
The same logic applies in the formula, we want to maximize $M$ this is done by maximizing the first term in the sum which the potential of links in the whole graph that are in community $c$, the second term is the fraction that would be expected to be in the community if the community was generated by a random model. If the first term is greater than the second this mean that there are more links in the community than expected hence the partition is good.

##### Question: Find communities in the network, and explain how you chose to identify the communities: Which algorithm did you use and how does it work?
###### Answer: We chose to use the Louvian-algorithm. The algorithm basically partitions the graph by creating a tree of the communities(dendrogram) found at each step of the algorithm, as you progress down the dendrogram the modularity increases. Each level/step is created using a heuristic which can be described as a decision-making rule for how to partition the graph. The operations performed at each level can be split into two phases the first phase is about optimizing the communities locally then the second step is about joining these small communities together to gives a larger community.

The algorithm requires an undirected version of the graph so we must convert the direct rapper network to an undirected graph

##### Question: Compute the value of modularity with the partition created by the algorithm.

This is a function for calculating the modularity of a single community

```python
def M_c(L_c, k_c, L):
    return (L_c/L)-(k_c/(2*L))**2
```

###### Answer: Here we use the previous function to calculate the total modularity of the network which is found to be 0.5447. We multiply the number of edges by to two obtain the sum of degrees, this is in accordance with the handshake lemma which states that $ k_c = 2L_c $

```python
M = 0
for graph in communities_sub_graphs:
    M += M_c(len(graph.edges), 2*len(graph.edges), len(G.to_undirected().edges))
M
```

##### Question: Calculate and store sentiment for every rapper

Before executing this line of code we removed some metadata in the first two lines of the document, since each feature is seperated by a tab we use `\t` as our delimiter, then we use the word column as index to support easy lookup access to the data associated with each word

```python
import pandas as pd
df = pd.read_csv("Data_Set_S1.txt", delimiter="\t")
df.set_index("word", inplace=True)
```

Here we get all the words are in the dataset

```python
rated_words = df.index.values.tolist()
```

This is the function we use to calculate the sentiment of a text. We start by creating a frequency distribution which we need later, then we iterate over all the rated words and calculate the sum in the numerator and the denominator of the formula for sentiment of a text which. At the end we perform the division.

```python
def calculate_sentiment(text):
    frequencies = nltk.FreqDist(text)
    sum_v_k_f_k = 0
    sum_f_k = 0
    for word in rated_words:
        frequency = frequencies[word]
        sum_v_k_f_k += df.loc[word]["happiness_average"] * frequency
        sum_f_k += frequency
    #Make sure we do not get a divide by 0 error
    if sum_f_k == 0:
        return 0
    return sum_v_k_f_k / float(sum_f_k)

```

Here we perform the sentiment calculation for each text

```python
west_sentiments = [(rapper[0], calculate_sentiment(rapper[1])) for rapper in west_rapper_tokens]
east_sentiments = [(rapper[0], calculate_sentiment(rapper[1])) for rapper in east_rapper_tokens]
```
