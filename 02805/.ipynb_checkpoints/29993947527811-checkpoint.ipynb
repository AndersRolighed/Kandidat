## Part 0: Building the network

> Creating the rapper network

**Reading the `.csv` files**

```python
# read the artists csv file (given in week04)
west_list = pd.read_csv('./WestCoastRappers.csv', index_col=0)
east_list = pd.read_csv('./EastCoastRappers.csv', index_col=0)
```

**Cleaning the data**

- remove artists that are in both coasts to avoid double-counting and ensure clear coast association for each artist

```python
# verify which artists are in both coasts
common_artists = set(west_list['WikipediaPageName']).intersection(set(east_list['WikipediaPageName']))

# remove these artists from both DataFrames
west_list = west_list[~west_list['WikipediaPageName'].isin(common_artists)]
east_list = east_list[~east_list['WikipediaPageName'].isin(common_artists)]

# concatenate the cleaned DataFrames
artist_list = pd.concat([east_list['WikipediaPageName'], west_list['WikipediaPageName']]).reset_index(drop=True)

print("Number of distict artists: ", len(artist_list))
```

**API Call**

```python
# url for the api
baseurl = "https://en.wikipedia.org/w/api.php?"

# dictionary to save the artists and their connections
link_dict = {}
text_dict = {}

# using a session for better performance with multiple requests
with requests.Session() as session:
    for artist in artist_list:
        params = {
            "action": "query",
            "prop": "revisions",
            "rvprop": "content",
            "format": "json",
            "titles": artist
        }
        
        try:
            response = session.get(baseurl, params=params)
             # check if request was successful
            response.raise_for_status()
            
            data = response.json()
            page_id = list(data['query']['pages'].keys())[0]
            
            # check if 'revisions' key exists
            if 'revisions' in data['query']['pages'][page_id]:
                wikitext = data['query']['pages'][page_id]['revisions'][0]['*']
                links = re.findall(r'\[\[([^\]|]+)(?:\|[^]]+)?\]\]', wikitext)
                # remove duplicate links
                links = list(set(links))  
                link_dict[artist] = links

            else:
                link_dict[artist] = []
            text_dict[artist] = wikitext
        
        except requests.RequestException as e:
            print(f"Error fetching data for {artist}. Error: {e}")
            # set empty list for artists with errors
            link_dict[artist] = [] 
```

**Filtering the Links**

```python
# get the names as a set instead of a dataframe
artist_set = set(artist_list)

filt_link_dict = {}
for key, values in link_dict.items():
    new_values = [value for value in values if value in artist_set]
    if new_values:
        filt_link_dict[key] = new_values
```

**Creating the Network**

```python
# initialize the directed graph
G = nx.DiGraph()

# adding nodes of the west coast artists
for artist in west_list['WikipediaPageName']:
    G.add_node(artist, coast='West')

# adding nodes of the east coast artists
for artist in east_list['WikipediaPageName']:
    G.add_node(artist, coast='East')

# adding edges based on the filtered links
for artist, links in filt_link_dict.items():
    for link in links:
        G.add_edge(artist, link)
```

**Visualizing the network**

```python
# positioning
pos = nx.spring_layout(G)

# assign node sizes based on degree (multiplied for better visualization)
degrees = [G.degree(n) * 50 for n in G.nodes()]

# define node colors based on 'coast' attribute
colors = ["blue" if G.nodes[node]["coast"] == "East" else "red" for node in G.nodes()]

# draw the graph
plt.figure(figsize=(8, 8))
plt.suptitle("Figure 1", fontsize=16, fontweight="bold")
plt.title("Rapper Network", fontsize=14)
nx.draw(G, pos, node_size=degrees, node_color=colors, alpha=0.7, edge_color='gray', width=0.5)
east_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='East Coast')
west_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='West Coast')
plt.legend(handles=[east_patch, west_patch], loc='upper right')
plt.show()
```

> Explain the strategy you have used to extract the hyperlinks from the Wiki-pages, assuming that you have already collected the rapper pages with the Wikipedia API.

Here's a step-by-step breakdown of the approach:

1. **API Calls**
    - we initiated a session to make API requests to Wikipedia for every artist in both `.csv`, specifying that we wanted the content (`wikitext`) through the parameter `rvprop` set to `content`;
    - by passing the artist Wikipedia page (artist name) in the `titles` parameter, we were able to retrieve the `wikitext` of the page.
2. **Error Handling**
    - to ensure potential issues like request failures we used `try` and `except`;
    - if there is an error with the API request, the artist would be logged and assigned an empty list.
3. **Hyperlink Extraction**
    - with the `wikitext` in hand, we used a specific regular expression pattern to extract hyperlinks;
    - this was crafted to match Wikipedia's link format;
    - the details on this pattern will be explained in the question below.
4. **Filtering Links**
    - since not all links extracted are relevant to our network, we filtered out the links that did not match with the artist names in our initial list;
    - this ensured that our network showed only connections between artists.
5. **Constructing the Network**
    - with the filtered links and artist names, we constructed a directed graph where nodes represent artists, and edges represent hyperlinks from one artist's page to another;
    - by using the `networkx` library, we also associated to each artist their respective coast (East or West).
6. **Visualization**
    - the network nodes were color-coded based on the coast association;
    - the size of each node was determined by the degree it had, allowing to distinct rappers who had more connections than others.

> Show the regular expressions you have built and explain in details how they work.

**Regular Expression**: `r'\[\[([^\]|]+)(?:\|[^]]+)?\]\]`

Purpose: extract hyperlinks from Wikipedia wikitext, accommodating both simple links (e.g., `[[Link]]`) and links with display text (e.g., `[[Link|Displayed Text]]`).

Part by part:
- `r`: denotes a "raw string" in Python, ensuring that the backslashes are treated as literal chars, making the regex pattern simpler.
- `\[\[`: matches the opening [[.
- `([^\]|]+)`: captures the main link. It grabs characters until it hits a | or ].
- `(?:\|[^]]+)?`: an optional section that matches optional display text after a | and before the closing ]] in links.
- `\]\]`: matches the closing ]].

## Part 2: Word-clouds

> Create a word-cloud for each coast according to the novel TF-TR method. Feel free to make it as fancy as you like.

**Getting the Rappers Pages**

```python
# convert each dataframes to a set of names
west_set = set(west_list['WikipediaPageName'])
east_set = set(east_list['WikipediaPageName'])

def fetch_text(artist_set): 
    text_dict = {}
    with requests.Session() as session:
        # iterate over the artists
        for artist in artist_set:
            # parameters for the request
            params = {
            "action": "query",
            "prop": "extracts",
            "exlimit": "1",
            "explaintext": "1",
            "format": "json",
            "titles": artist
            }
            
            try:
                response = session.get(baseurl, params=params)
                # check if request was successful
                response.raise_for_status()
                
                # parse the response
                data = response.json()
                page_id = list(data['query']['pages'].keys())[0]
                
                # extract the text content from the 'extract' property
                extract = data['query']['pages'][page_id].get('extract', "")
                
                # store the extract in the dictionary
                text_dict[artist] = extract
            
            # handle exceptions in case of request errors
            except requests.RequestException as e:
                print(f"Error fetching data for {artist}. Error: {e}")
                
                # assign an empty string for artists with errors
                text_dict[artist] = ""
        
        return text_dict

west_text_dict = fetch_text(west_set)
east_text_dict = fetch_text(east_set)
```

**Tokenize**

```python
# punctuation and stopwords
punctuation = string.punctuation + "``" + "''" + "-" + "" + "--" + "==" + "===" + "'s"
stopwords = set(nltk.corpus.stopwords.words("english"))

# tokenize and clean text from a dictionary
def tokenize_dict_content(text_dict):
    # create an empty list to store tokenized words
    tokenized_words = []
    
    # lemmatizer
    lemmatizer = nltk.WordNetLemmatizer()
    
    for text in text_dict.values():
        # tokenize text
        tokens = nltk.word_tokenize(text)
        
        # remove punctuation and stopwords, set to lowercase, and lemmatize words
        tokens = [
            lemmatizer.lemmatize(word.lower())
            for word in tokens
            if word not in punctuation and word.lower() not in stopwords
        ]
        
        # append tokens to list
        tokenized_words.extend(tokens)
    
    return tokenized_words

# tokenize for each coast
east_tokens = tokenize_dict_content(east_text_dict)
west_tokens = tokenize_dict_content(west_text_dict)
```

**Calculate TF**

```python
def calculate_tf(tokens):
    # frequency distribution of tokens
    tf = nltk.FreqDist(tokens)
    return tf

east_coast_tf = calculate_tf(east_tokens)
west_coast_tf = calculate_tf(west_tokens)
```

**Calculate TR**

```python
def calculate_tr(tf1, tf2, c=1.0):
    # calculate Term Ratios (TR) for all tokens in tf1
    tr = {token: tf1[token] / (tf2[token] + c) for token in tf1}
    return tr

# constants and term ratios
c = 4.0
east_coast_tr = calculate_tr(east_coast_tf, west_coast_tf, c)
west_coast_tr = calculate_tr(west_coast_tf, east_coast_tf, c)
```

**Calculate Score**

```python
def calculate_score(tf, tr):
    # calculate score for each token as the product of its TF and TR
    score = {token: tf[token] * tr[token] for token in tf}
    return score

east_coast_score = calculate_score(east_coast_tf, east_coast_tr)
west_coast_score = calculate_score(west_coast_tf, west_coast_tr)

# sort tokens by Score to find the most distinctive words
east_coast_top = sorted(east_coast_score.items(), key=lambda x: x[1], reverse=True)[:5]
west_coast_top = sorted(west_coast_score.items(), key=lambda x: x[1], reverse=True)[:5]
```

>  Explain your process and comment on your results.

1. **Data Collection**
    - for each rapper's Wikipedia page, we fetched its content using the Wikipedia API, following the same structure we used before.
2. **Tokenization**
    - we tokenized the each text for each cost, breaking the text into individual words and tokens;
    - during this tep, we removed common `stopwords` and `ponctuation` to ensure the data is clean and representative of meaningful terms;
    - we also perform lemmatization, which means reducing the words to their *base form* (e.g., "running" becomes "run").
3. **Term Frequency (TF)**
    - we calculated the TF for each token, which represents how often a word appears in the text for each coast (using the built-in `FreqDist` from `nltk`).
4. **Term Ratio (TR)**
    - To determine the uniqueness of a word for each coast, we used the concept of Term Ratios (TR). The formulas used were:
    $$
    w_{t}^{(w)} = \frac{\text{TF}^{(w)}_t}{\text{TF}^{(e)}_t + c} \hspace{1cm} \text{and} \hspace{1cm} w_{t}^{(e)} = \frac{\text{TF}^{(e)}_t}{\text{TF}^{(w)}_t + c}
    $$
    - this calculates a weight for each token based on how frequent it appears in the other coast;

## Part 3: Communities

> In your own words, explain what the measure "modularity" is, and the intuition behind the formula you use to compute it.

**Modularity**

Modularity is a measure of the structure of a graph, measuring the density of connections within a module or community. Essentially, it helps in understanding how well-separated or distinct the communities within a network are.

**Intuition behind the formula**

Formula:
$Q = \frac{1}{2m} \sum_{ij} \left( A_{ij} - \frac{k_i k_j}{2m} \right) \delta(c_i, c_j)$

Where:
- $m$ is the total number of edges in the network.
- $A_{ij}$ is an element of the adjacency matrix, which is 1 if there's an edge between node $(i)$ and node $j$, and 0 otherwise.
- $k_i$ and $k_j$ are the degrees of node $i$ and $j$ respectively.
- $\delta(c_i, c_j)$ is the Kronecker delta function, which is 1 if nodes $i$ and $j$ are in the same community and 0 otherwise.

The intuition that stands behind the modularity formula. Let's imagine a network where edges between nodes are placed entirely at random, without any preference. In such a case, there would be some probability that a link would exist between any two nodes. Then on the other hand, in an actual network if the communities detected are meaningful, there should be more edges between nodes within the same community as what you'd expect from the random scenario. Modularity measures the difference between these two scenarios: the actual number of intra-community edges minus the expected number of such edges in the random scenario.

> Find communities in the network, and explain how you chose to identify the communities: Which algorithm did you use, and how does it work?

To identify the communities in the network, **The Louvain Algorithm** has been used. The working process of this algorithm can be split into a couple of phases.

1. **Initialization**
    - each node in the network is treated as its community;
    - at the beginning, for a graph with N nodes, there are N communities.
2. **Modularity Optimization**
    - for each node in the network, the method evaluates the change in modularity if the node is removed from its current community and placed in a neighbouring community;
    - the node is then assigned to the community where this change is maximal, but only if this change yields an increase in the overall modularity;
    - this process is repeated for all nodes until no increase in modularity is possible. 
3. **Community Aggregation**
    - after the above step, communities that have been detected are aggregated into "super-nodes" to form a new network;
    - links between nodes in the same community are turned into self-loops, and links between nodes in different communities are summarized as weighted links between the respective communities. 
4. **Iteration**
    - steps 2 and 3 are iteratively repeated on the resulting aggregated network until there's no further improvement in modularity.

```python
# helper function to split given dictionary into a set of nodes
def partition_to_communities(partition_dict):
    # convert a node: community dictionary into a list of sets of nodes
    communities = defaultdict(set)
    for node, community_id in partition_dict.items():
        communities[community_id].add(node)
    return list(communities.values())
```

## Part 4: Sentiment of Communities

> Average the average sentiment of the nodes in each community to find a community-level sentiment.

```python
# load in LabMT word list

labMT_word_list = {}
with open("Data_Set_S1.txt", 'r') as file:
    for line in file:
        values = line.strip().split("\t")
        labMT_word_list[values[0].lower()] = float(values[2].strip())
```

```python
# creating a function to calculate the sentiment
def calculate_sentiment(text):
    tokens = text.lower().split()
    tokens_size = 0
    sentiment = 0
    for token in tokens:
        if token in labMT_word_list:
            sentiment += labMT_word_list[token]
            tokens_size += 1
    if tokens_size == 0 or sentiment == 0:
        return None
    else:
        return sentiment/tokens_size


def calculate_coast_sentiment(coast):
    coast_sentiment = 0
    coast_count = 0
    for rapper in coast:
        sentiment_value = calculate_sentiment(rapper)
        if sentiment_value is not None:
            coast_sentiment += sentiment_value
            coast_count += 1

    return coast_sentiment/coast_count if coast_count != 0 else 0
```

```python
sentiment_scores = []
sentiment_dict = {}

for rapper, text in text_dict.items():
    sentiment_value = calculate_sentiment(text)
    if sentiment_value is not None:
        sentiment_dict[rapper] = sentiment_value
        sentiment_scores.append(sentiment_value)

# plotting
plt.figure(figsize=(10, 6))
plt.hist(sentiment_scores, bins=30, alpha=0.7, color='green')
plt.xlabel("Sentiment Score")
plt.ylabel("Number of Rappers")
plt.suptitle("Figure 7", fontsize=16, fontweight='bold')
plt.title("Distribution of Rapper Sentiment Scores", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

> Use the "label shuffling test" (Week 5 and 8) to test if the coast with the highest Wikipedia page sentiment has a page sentiment that is significantly higher (5% confidence bound) than a randomly selected group of rappers of the same size.

```python
two_community_partition[0]
```

```python
# since the West Coast has the happiest sentiment we will be using the West Coast for this test

permutations = 1000
extremes = 0

west_coast_sentiment = calculate_coast_sentiment(two_community_partition[1])
west_coast_rapper_count = len(two_community_partition[1])

all_rappers = list(two_community_partition[1]) + list(two_community_partition[0])

test_sentiments = []

for _ in range(permutations):
    # copy the list to avoid altering the original
    shuffled_rappers = all_rappers.copy()
    # shuffling
    random.shuffle(shuffled_rappers)
    
    selected_rappers = shuffled_rappers[:west_coast_rapper_count]

    # test sentiment
    test_sentiment = calculate_coast_sentiment(selected_rappers)

    # handle 0 sentiment
    if test_sentiment is not None and test_sentiment >= west_coast_sentiment:
        extremes += 1

#p-value
p  = extremes/permutations

#compare it to confidence level(alpha)
alpha = 0.05
print("Probability Value is:",p)
print("Alpha:",alpha)
```

> Does the result make sense to you? Elaborate.

1. **Probability Value (p-value) = 0.232**
    - represents the proportion of the permuted data that had a sentiment greater than or equal to the observed sentiment of the West Coast rappers;
    - a p-value of 0.232 suggests that in about 23.2% of the random permutations, a group of rappers had a sentiment score greater than or equal to the actual sentiment score of the West Coast rappers.
2. **Alpha ($\alpha$) = 0.05**
    - this is the chosen significance value;
    - it's a threshold below which we would reject the null hypothesis.
    
**Null Hypothesis**: the observed difference in sentiment between the coasts is due to random chance.

Given these values:
- **p > $\alpha$ (0.232 > 0.05)**, which means we fail to reject the null hypothesis;
- in other words, we don't have enough evidence to say that the West Coast's sentiment score is significantly higher than what you might get with a random group of rappers of the same size, at 5% significance level.
