# Part 0: Building the network 

The following section answers the questions:
> * Explain the strategy you have used to extract the hyperlinks from the Wiki-pages, assuming that you have already collected the rapper pages with the Wikipedia API.
> * Show the regular expressions you have built and explain in details how they work.


To create our network, we have developed four functions. These are:
- **get_wiki_text()**: This function takes a rapper name as input and outputs the wikipedia page for that rapper using the Wikipedia API.
- **extract_links()**: This function takes a Wikipedia text as input and outputs all the hyperlinks in that text.
- **rapper_check()**: This function takes a hyperlink string as an input and checks whether that string exists in our dataframe with  rappers.
- **build_network()**: This function take the rapper dataframe as input and outputs the network for that dataframe by combining the three aforementioned functions.

For the purpose of this assignment, we will only focus on the **extract_links()** function, as that is the one making use of regular expressions. However, for the sake of transparency, and to give the reader insight into how the functions tie together, we will provide the code for all four functions. The function using the Wikipedia API to retrieve the rapper pages can be seen below.

````python
def get_wiki_text(rapper):
    baseurl = "https://en.wikipedia.org/w/api.php?"
    params = {
    "action": "query",
    "prop": "revisions",
    "rvprop": "content",
    "format": "json",
    "titles": rapper
    }
    
    wikitext = requests.get(baseurl, params=params)
    data = wikitext.json()
    page = next(iter(data["query"]["pages"].values()))
    return page["revisions"][0]["*"]
````

After using this function to retrieve the text from a rapper's Wikipedia page, we now wanted to extract the hyperlinks. This is were we used regular expressions in the **extract_links()** function seen below. The function looks for links using regular expressions with the **re.findall()** command, and outputs them as strings in an array called "links".

````python
def extract_links(wikitext):
    links = re.findall(r"\[\[(.*?)\]\]", wikitext)
    return [link.split('|')[0] for link in links]
````

Let's dive deeper into how the regular expressions work. The function uses the regular expression ```r"\[\[(.*?)\]\]"``` to look for (or "match") hyperlinks in the text according to the double square brackets format: ```[[Link Title]]```.

The expressions works as follows:
- ```\[``` and ```\]``` represent ```[``` and ```]```. As these are considered special characters used for classes, we need to use ```\``` to turn them into literal characters. These are found twice to match the format explained above.
- ```(.*?)``` Is capturing the contents within the square brackets. This is called the "capturing group" and is denoted by the parenthesis ```()```. To break it down:
    - ```.``` matches any single character except newline characters.
    - ```*``` matches zero or more of the previous element (the ```.```), so ```.*``` together match as many characters as possible.
    - ```?``` then ensures that we avoid matching multiple links at once, by making the previous ```*``` term "non-greedy". This means it matches as few characters as possible until the next part of the pattern is found. If this term was not included, we would capture everything from the first ```[[``` to the last ```]]``` found, and hence with its inclusion we now only match one link at a time.

However, this regular expression was not quite sufficient, as sometimes we would see links where an additional display text is included in the link, after a ```|``` in the following format: ```[[WikipediaPageName|DisplayText]]```. This we wanted to fix, as it would mess up later connections between rappers.

To fix this, we want get rid of everything before and including ```|```. Now to do this, we could include an additional regular expressions, but instead we used a simple trick with the **.split()** function in the **return** statement. As seen in the function, the output is: ```return [link.split('|')[0] for link in links]```. Here, ```[link.split('|')[0] for link in links]``` splits any link at ```|``` if present, and includes only the first index ```[0]``` in our output, which is the actual hyperlink (or "WikipediaNamePage") that we want to use for our network in case it refers to another rapper's page.

We check if the link refers to another rapper in the function below, wherethe hyperlink string is compared to the rapper names in our rapper dataframe.

````python
def rapper_check(link, df):
    return link in df['WikipediaPageName'].values
````

Finally, the East Coast and West Coast datasets ```ec``` and ```wc``` are combined, and we run the last function ```build_network(```that ties everything together and builds the network as seen below. Also, self loops are removed as they are deemed unmeaningful for the purpose of this assignment.

````python
# merge eastcoast and westcoast dataframes into one
all_rappers = pd.concat([wc, ec])
````

````python
def build_network(df):
    G = nx.DiGraph()
    
    # iterate over all rappers
    for index, row in df.iterrows():
        rapper = row['WikipediaPageName']
        coast = 'West' if rapper in wc['WikipediaPageName'].values else 'East'
        
        # get wiki page contents
        wikitext = get_wiki_text(rapper)
        
        # retrieve all the links from page
        links = extract_links(wikitext)
        
        # add rapper as a node with coast as an attribute
        G.add_node(rapper, coast=coast)
        
        # add edge if link points to another rapper (for each link)
        for link in links:
            if rapper_check(link, df):
                G.add_edge(rapper, link)
                
    # discard nodes with no in- or out- degrees
    isolated_nodes = [node for node, degree in dict(G.degree()).items() if degree == 0]
    for node in isolated_nodes:
        G.remove_node(node)
        
    return G
````

````python
# build the network using the function above
network = build_network(all_rappers)
````

````python
# remove self loops
network.remove_edges_from(nx.selfloop_edges(network))
````

````python
network_ud = network.to_undirected()
````

````python
def visualize_network(G):
    plt.figure(figsize=(20, 20))

    pos = nx.kamada_kawai_layout(G)

    # make westcoast nodes red and eastcoast nodes blue
    node_colors = ['red' if data['coast'] == 'West' else 'blue' for _, data in G.nodes(data=True)]
    node_sizes = [10 * G.degree(node) for node in G.nodes]
    nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=node_sizes ,width=0.5, edge_color="gray")

    plt.title("Rapper Network")
    plt.axis('off')
    plt.show()

visualize_network(network_ud)
````

Create your own version of the word-clouds (from lecture 7). For this exercise we assume you know how to download and clean text from rappers' Wikipedia pages.

First we defined the function "extract_text()" in order to iterate trough each .txt file of each coast of rappers. Then we tokenised all of the text into words, removed punctuation and set all letters to lower case.

````python
import os
import string

def extract_text(directory_path):
    all_words = []

    # Iterate through each file in the directory
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        if filename.endswith(".txt") and os.path.isfile(file_path):
            with open(file_path, 'r', encoding='utf-8') as file:
                text_data = file.read()

                # Tokenize the text into words, remove punctuation, and set to lowercase
                words = text_data.split()
                words = [word.strip(string.punctuation).lower() for word in words]

                # Extend the list of words
                all_words.extend(words)

    return all_words

# Define the paths for East Coast and West Coast directories
east_coast_path = "/Users/dimi/Machine Learning/Social Graphs/texts/eastcoasttexts"
west_coast_path = "/Users/dimi/Machine Learning/Social Graphs/texts/westcoasttexts"

# Process East Coast and West Coast text data
east_coast_words = extract_text(east_coast_path)
west_coast_words = extract_text(west_coast_path)
````

Secondly, we created two lists with the counts of each word in both coasts.

````python
# Create TF (Term Frequency) List
east_coast_tf = Counter(east_coast_words)
west_coast_tf = Counter(west_coast_words)
````

````python
west_coast_tf
````

After, we created a TF-TR list where we calculated the weight and the frequency of each word from both coasts.

````python
# Term Frequency - term Ratio Lists
c = 8

# Create TF-TR lists for East Coast
east_coast_tftr = {}
for word, frequency in east_coast_tf.items():
    west_coast_frequency = west_coast_tf[word] if word in west_coast_tf else 0
    weight_east = frequency / (west_coast_frequency + c)
    east_coast_tftr[word] = frequency * weight_east

# Create TF-TR lists for West Coast
west_coast_tftr = {}
for word, frequency in west_coast_tf.items():
    east_coast_frequency = east_coast_tf[word] if word in east_coast_tf else 0
    weight_west = frequency / (east_coast_frequency + c)
    west_coast_tftr[word] = frequency * weight_west
````

In order to create the word clouds, we first had to convert the two TF-TR lists from above into single strings. This way we then generate the word clouds for both coasts. We create an object of class "WordCloud()" and call the "generate()" method. This way we created a Word Cloud for both coasts.

````python
# Take out "e", but stands for e-40???
w = WordCloud()
stop_words = list(w.stopwords)
custom_stop_words = ['e']
stop_words = set(stop_words + custom_stop_words)
````

````python
# Convert TF-TR lists to a single string
east_coast_string = " ".join([word for word, score in east_coast_tftr.items() for _ in range(round(score))])
west_coast_string = " ".join([word for word, score in west_coast_tftr.items() for _ in range(round(score))])

# Create a word cloud for the East Coast
east_coast_wordcloud = WordCloud(collocations=False, background_color='black', width=1600,height=800, stopwords=stop_words).generate(east_coast_string)

# Create a word cloud for the West Coast
west_coast_wordcloud = WordCloud(collocations=False, background_color='black', width=1600,height=800, stopwords=stop_words).generate(west_coast_string)
````

````python
# Display the word clouds
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.imshow(east_coast_wordcloud, interpolation='bilinear')
plt.title("East Coast")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(west_coast_wordcloud, interpolation='bilinear')
plt.title("West Coast")
plt.axis("off")

plt.show()
````

We iterate over the word cloud lists and print the top 5 words for each coast. Since there are a lot of stop words in this case, we decided to create a list of common stop words. Which we then filter out of the system.

````python
# Common stopwords to filter out
common_stopwords = set(['he','on','his','was','the', 'and', 'in', 'a', 'is', 'it', 'to', 'of', 'this', 'that', 'with', 'for', 'as'])

# Define a function to filter out stopwords and non-word tokens
def filter_words(tftr_scores, coast_name):
    meaningful_words = [(word, tftr_score) for word, tftr_score in tftr_scores.items() if word not in common_stopwords and word.isalpha()]
    sorted_words = sorted(meaningful_words, key=lambda x: x[1], reverse=True)
    
    print(f"Top 5 meaningful words for {coast_name}:")
    for word, tftr_score in sorted_words[:5]:
        print(f"{word}: TF-TR = {tftr_score}")
````

# Part 3: Communities

Find communities and their modularity (from lecture 7).

Here's what you need to do:
> * In your own words, explain what the measure "modularity" is, and the intuition behind the formula you use to compute it. 

Modularity describes a measure of a network's or graph's structure which quantifies how strongly a network is divided into modules, also known as groups, clusters, or communities. Dense connections exist between nodes inside modules in networks with strong modularity, yet sparse connections exist between nodes in different modules. Positive modularity values indicate a good division of the network into communities.

This calculation's intuition is based on comparing a community's actual connections to those that one would anticipate in a random network. A split into communities that is superior to randomness is indicated by a positive modularity score. Stronger community structures inside the network are indicated by higher modularity levels.

First we split the rappers network into two parts. One for the east coast's and one for the west coast's rappers. The reason being the strong split between the coasts.

````python
# Define West Coast and East Coast nodes
west_coast_nodes = [node for node, data in network_ud.nodes(data=True) if data['coast'] == 'West']
east_coast_nodes = [node for node, data in network_ud.nodes(data=True) if data['coast'] == 'East']

# Create the partition
community_partition = {node: 0 for node in west_coast_nodes}
community_partition.update({node: 1 for node in east_coast_nodes})

print("Number of Nodes:")
print("West Coast:",len(west_coast_nodes))
print("East Coast", len(east_coast_nodes))
````

````python
# Create separate networks for East Coast and West Coast
network_east_coast = network.subgraph(east_coast_nodes).copy().to_undirected()
network_west_coast = network.subgraph(west_coast_nodes).copy().to_undirected()
````

````python
def louvain_alg(network):
    # compute the best partition
    partition = community_louvain.best_partition(network)

    # draw
    size = float(len(set(partition.values())))
    pos = nx.spring_layout(network)
    count = 0.
    for com in set(partition.values()) :
        count = count + 1.
        list_nodes = [nodes for nodes in partition.keys()
                                    if partition[nodes] == com]
        nx.draw_networkx_nodes(network, pos, list_nodes, node_size = 20,
                                        node_color = str(count / size))
        nx.draw_networkx_edges(network,pos, alpha=0.5)
    plt.show()
````

````python
louvain_alg(network_east_coast)
````

````python
louvain_alg(network_west_coast)
````
